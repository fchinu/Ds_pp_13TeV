```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append('/home/fchinu/Run3/ThesisUtils')
from DfUtils import apply_model_in_batches

from hipe4ml.model_handler import ModelHandler
```

## 2 < pT < 2.5

```python
ModelHandl = ModelHandler()
ModelHandl.load_model_handler('/home/fchinu/Run3/Ds_pp_13TeV/ML/Training/pt2_2.5/ModelHandler_pT_2_2.5.pickle')
```

```python
df = apply_model_in_batches(ModelHandl, ["fM", "fSign"], "/home/fchinu/Run3/Ds_pp_13TeV/Datasets/Ds_pp_run3_ml/Train179897/LHC22o_pass6_small.parquet", "2<fPt<2.5")
dfSelPos = df.query("ML_output_Bkg < 0.01 and fSign > 0")
dfSelNeg = df.query("ML_output_Bkg < 0.01 and fSign < 0")

counts,bin_edges = np.histogram(dfSelPos['fM'],bins=75, range=(1.7, 2.1))
plt.errorbar(x=bin_edges[1:], y=counts, yerr=np.sqrt(counts), fmt='o', label='$D_s^+$')

counts,bin_edges = np.histogram(dfSelNeg['fM'],bins=75, range=(1.7, 2.1))
plt.errorbar(x=bin_edges[1:], y=counts, yerr=np.sqrt(counts), fmt='o', label='$D_s^-$')

plt.legend()
```
